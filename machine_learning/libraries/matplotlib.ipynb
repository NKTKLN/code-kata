### üß† Implementing Custom Model Algorithms

```python
x = np.arange(0, 20, 0.1)
```

```python
plt.plot(x, x**2, "r--", linewidth=3)
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.title("y=x^2")
plt.grid(linestyle="-.")
plt.show()
```

```python
plt.plot(x, np.sin(x), "g:", label="sin(x)")
plt.plot(x, np.cos(x), "b-.", label="cos(x)")
plt.title("sin & cos")
plt.legend(loc="upper right")
plt.axis((-1, 20, -1.5, 1.5))
x_ticks = np.arange(0, 20, np.pi)
plt.xticks(x_ticks, [f"{i}œÄ" if i else "0" for i in range(len(x_ticks))])
plt.show()
```

```python
t = np.arange(0, 5, 0.2)

for i, s in enumerate(["r-.", "g^", "b.", "ms", "yx"]):
    plt.plot(t, np.full_like(t, i), s)

plt.show()
```

```python
species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    "Bill Depth": (18.35, 18.43, 14.98),
    "Bill Length": (38.79, 48.83, 47.50),
    "Flipper Length": (189.95, 195.82, 217.19),
}

x_i = np.arange(len(species))
width = 0.25

for multiplier, (attribute, measurement) in enumerate(penguin_means.items()):
    plt.bar(x_i + width * multiplier, measurement, width, label=attribute)

plt.ylabel("Length (mm)")
plt.title("Penguin attributes by species")
plt.xticks(x_i + width, species)
plt.legend(loc="upper left", ncols=3)
plt.ylim(0, 250)
plt.show()
```

```python
n = 50
xs = np.random.uniform(23, 32, n)
ys = np.random.uniform(0, 100, n)
zs = np.random.uniform(-50, -25, n)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(xs, ys, zs, marker="x")

plt.show()
```
